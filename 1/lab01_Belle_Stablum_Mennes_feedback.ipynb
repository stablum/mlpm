```python
%pylab inline

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix of shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X, showAxes=False):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        # see http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot
        # subplot params: numrows, numcols, fignum where fignum ranges from 1 to numrows*numcols.
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        if not showAxes:
            ax.set_xticks([])
            ax.set_yticks([])
```

At first we need a dataset $\mathbf{D}_1$

Then the vector of means $\mathbf{\mu}$ has to be obtained, in order to be subtracted afterwards before the calculation of the covariance matrix.

```python
mu = np.array([ np.mean(d) for d in D1 ])
mu
```

A new dataset $\mathbf{D}_2$ is obtained subtracting each row of $\mathbf{D}_1$ with the mean vector $\mu$

```python
D2 = (D1.T - mu).T
print D2
print np.mean(D2)
```

The dataset $\mathbf{D}_2$ represents some observations of a $\mathbf{x}$ aleatory vector, which has one observation per row.
Now it's important to calculate the covariance matrix.
$$\mathbf{\Sigma} = \mathbb{E}(\mathbf{xx}^\mathrm{T})$$

```python
Sigma = np.cov(D2)
print Sigma
print Sigma.shape

```

Recall that $\Sigma \Phi = \Phi \Lambda$, where $\Lambda$ is the diagonal matrix of eigenvalues.


Therefore the matrix of eigenvectors $\Phi$ has to be found, because this matrix will diagonalize the covariance matrix of  $\mathbf{D}_2$.

```python
lambdas, Phi = np.linalg.eig(Sigma)

print lambdas
print Phi
```

$\Phi^\mathrm{T}$ is obtained

```python
Phi_T = np.transpose(Phi)
Phi_T
```

Because $\Phi^\mathrm{T} \Sigma \Phi = \Lambda$ and  $\Sigma \Phi = \Phi \Lambda$, it holds that $\Sigma \mathbf{x} = \Phi \Lambda \Phi^{\mathrm{T}} \mathbf{x} $. This says that the effect of $\Sigma$ on $\mathbf{x}$ is the same as the effect of $\Lambda$ on $\mathbf{y}$ in the transformed space defined by $\mathbf{y} = \Phi^\mathrm{T} \mathbf{x}$.

So at this point, all data needs to be processed trough $\Phi^\mathrm{T}$ in order to find the transformed data $\mathbf{y}$

```python
ys = np.dot(Phi_T, D2)
ys
```

Sanity check for correctness of decorrelation:

```python
cov_ys  = np.cov(ys)
np.set_printoptions(suppress=True)
print cov_ys

ax = imshow(log(abs(cov_ys)), cmap='gray', interpolation='nearest')
```

From this we can infer that $\Lambda^{-\frac{1}{2}} \Phi^{\mathrm{T}} \mathbf{x}$ will yield the whitened data.

### 1.4.2 Resulting functions

 Now we know enough to define the necessary functions: `decorrelation}(X)`, that takes the raw data matrix as argument; and `whitening(X)`, that takes the decorrelated data as argument. These can successively be applied to yield the final whitening results.

```python
def decorrelation(data):
    mu = np.array([ np.mean(d) for d in data ]) # calculating mean
    data_shifted = (data.T - mu).T # removing mean
    Sigma = np.cov(data) # covariance matrix
    lambdas, Phi = np.linalg.eig(Sigma) # calculating Phi (eigenvectors)
    Phi_T = np.transpose(Phi) # transpose of Phi
    ys = np.dot(Phi_T, data_shifted) # decorrelated data
    return ys
```

```python
def whitening(decorrelated_data):
    Lambda = np.cov(decorrelated_data)
    eigenvalues = np.diag(Lambda)
    Lambda12 = np.diag(1./sqrt(eigenvalues+0.0001))
    whitened = np.dot(Lambda12, decorrelated_data)
    return whitened
```

```python
## convenience function
## returns whitened data
# additionally can print  data  preprocessing steps reuslts
def getPreprocessedData(X, printStuff=True):    
    if printStuff:
        print "Raw data covariance:\n%s\n" % np.cov(X)
        
    Xdecorrelated = decorrelation(X)
    if printStuff:
        print "Decorrelated data covariance:\n%s\n" %  np.cov(Xdecorrelated)
        
    Xwhitened = whitening(Xdecorrelated)
    if printStuff:
        print "Whitened data covariance:\n%s\n" % np.cov(Xwhitened)

    return Xwhitened
```

## 1.6 Covariant algorithm

A covariant algorithm is an algorithm which gives the same results independent on the units in which quantities are measured.

For example the steepest descent rule is a prime example of a non-covariant algorithm. A dimensionless objective function $ L ( \mathbf{w} ) $ is defined, its derivative with respect to $ \mathbf{w} $ is in order to update $ \mathbf{x} $ proportionally to a constant learning rate $\eta$ :

$$ \Delta w_i = \eta \frac{\partial L}{\partial w_i} $$

The behavior of the steepest descent learning algorithm is not covariant with respect of rescaling of the vector $\mathbf{w}$.

## 1.7 Independent Component Analysis

Implementation of the covariant ICA function as outlined by McKay (2003).

```python
### McKay approach with covariant
def ICA_McKay_covariant(X, activation_function, learningRate, maxIterations, epsilon, printInterval):    
    
    W =  random_nonsingular_matrix(num_sources)

    deltaW = W*sys.maxint    
    
    nrInstances = X.shape[1]

    iteration = 0
    while mean(abs(deltaW)) > epsilon and iteration < maxIterations:       
        
        ## A <-- current source approximation
        # multiply by column!
        #
        # [X * column1(W); X*column2(W)]
        # equals
        # X * W
        A = np.dot(W,X)
        
        ## Z <-- map activation function on A
        Z =  np.array(map(activation_function, A))       

        ## Xprime <-- map activation function on A
        Xprime = dot(W.T, A) 
        
        ## take the average of the gradient update rule for all column 
        ## vectors of X
        #
        # dot(Z,Xprime.T) = get A back through W for maintaining covariance 
        averageGradient = W + (dot(Z,Xprime.T) / nrInstances)            
                
        ## update W
        deltaW = learningRate * averageGradient
        W = W + deltaW
       
        
        
        iteration += 1
        if iteration % printInterval == 0:
            print "Iteration: %d" % iteration

    output = "Stopped at iteration %d" % iteration
    if iteration >= maxIterations:
        output += " because max nr. of iterations (%d) exceeded." %  maxIterations
    else:
        output += " because mean(abs(deltaW)) >= epsilon (%2.8f)" % epsilon
    print output
    
    Sapprox = dot(W,X)  

    return Sapprox


```

Let the function with the specific name `ICA(X, activation_function, learning_rate)` refer to the relevant ICA implementation

```python
###
# Main ICA function
###
# which ICA function to use when calling ``ICA(X, activation_function, learningRate)''
def ICA(X, activation_function, learningRate, maxIterations=10000, epsilon=0.0001, printInterval=100):
    print "Going to run ICA."
    return ICA_McKay_covariant(X, activation_function, learningRate, maxIterations, epsilon,printInterval)
```

This runs the ICA algorithm on the artificial data using each activation function.
The output consists of the sum of squared differences (SSD) with the source signals
since in the source approximation the signals can have a different order this difference is calculated for 
each permutation of the approximation and the minimum is the final result

First we define a function to calculate the difference between source to be approximated and the  approximation of the source 
. We use the sum of squared differences as distance  metric
 because orderings of signals can differ in the approximated result,
 the returned value is implemented as the minimum difference value over all
 permutations (of signal orderings)

```python

import itertools

def calcDifference(approximation, source_2b_appr):
    
    perms = list(itertools.permutations(range(0,num_sources),num_sources))
    
    min_diff = float("inf")
    min_perm = []    
    
    print len(perms)
    
    for p in perms:
        dif = 0
        for i in range(num_sources):
            source = source_2b_appr[i,:]
            appr = approximation[p[i],:] 
            
            dif_dim = sum(np.power(source-appr, 2)) 
            #print dif_dim
            dif += dif_dim
            
        if dif<min_diff:
            min_diff = dif
            min_perm = p
            
    return (min_diff, min_perm)
```

## 1.10 Excess Kurtosis

###  Analytical derivation of the kurtosis of the four priors that we derived from the activation functions before:

<h3>$k^\mathrm{th}$ standardized moment</h3>
In probability, the $k^\mathrm{th}$ standardized moment of a probability distribution is defined as:
$$\frac{\mu_k}{\sigma^k}$$
where $\mu_k$ is the $k^\mathrm{th}$ moment about the mean:
$$\mu_k = \operatorname{E} \left[ ( X - \operatorname{E}[X] )^k \right]  = \int_{-\infty}^{+\infty} (x - \mu)^k f(x)\,dx.$$
and $\sigma$ is the standard deviation.
<h3>Fourth standardized moment</h3>
$${\beta_2=}\frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} {=} \frac{\mu_4}{\sigma^4}$$
<h3>Excess Kurtosis</h3>
Excess Kurtosis is the measure of the peakedness of a probability density distribution, defined as the 
$$\gamma_2 = \beta_2 - 3 = \frac{\mu_4}{\sigma^4} - 3$$
The $-3$ addend has the purpose of imposing excess kurtosis as $0$ in the normal distribution.

<h3>Excess kurtosis for $\phi(a) = -\verb+tanh+(a)$</h3>
the probability density function of the activation function $\phi(a) = -\verb+tanh+(a)$ is:
$$e^{ \int -\verb+tanh+(a) \mathrm{d}a} = \verb+sech+(a) = \frac{2}{e^x - e^{-x}}$$

Since $\int_{-\infty}^{+\infty} \verb+sech+(x) \mathrm{d}x = \pi$, the probability density function has to be normalized:
$$p(x) = \frac{1}{\pi}\verb+sech+(x)$$

Its mean is:
$$\mu = \int x p(x) \mathrm{d}x = \int x \frac{2}{e^x + e^{-x}} \mathrm{d}x $$

The mean is $0$ since the formula $h(x)$ within the integral is symmetric for both axis :
$$ -h(-x) = - (-x) \frac{2}{e^{-x} + e^x} = x \frac{2}{e^x + e^{-x}} = h(x)$$

Its excess kurtosis is therefore:
$$
\frac{\mu_4}{\sigma^4} - 3 
= \frac{\int_{-\infty}^{+\infty} (x - \mu)^4 p(x)\,\mathrm{d}x}{\sigma^4} -3
= \frac{\int_{-\infty}^{+\infty} x^4 p(x)\,\mathrm{d}x}{\Big(\big(\int_{-\infty}^{+\infty} (x-\mu)^2p(x)\mathrm{d}x \big)^\frac{1}{2}\Big)^4} -3
= \frac{\int_{-\infty}^{+\infty} x^4 p(x)\,\mathrm{d}x}{\Big(\int_{-\infty}^{+\infty} x^2p(x)\mathrm{d}x \Big)^2} -3
$$

Using the integration by parts, we get:
$$
\int_{-\infty}^{+\infty} x^4 p(x)\,\mathrm{d}x
= \big[ x^4 \big]_{-\infty}^{+\infty} - \int_{-\infty}^{+\infty}4x^3\mathrm{d}x 
= \big[ x^4 \big]_{-\infty}^{+\infty}
$$

having eliminated $\int_{-\infty}^{+\infty}4x^3\mathrm{d}x $ because integral of a symmetrical function on both axes.

and:

$$
\Bigg(\int_{-\infty}^{+\infty} x^2p(x)\mathrm{d}x \Bigg)^2
= \Bigg( 
\big[ x^2 \big]_{-\infty}^{+\infty} - \int_{-\infty}^{+\infty}2x\mathrm{d}x 
\Bigg)^2
= \Bigg( 
\big[ x^2 \big]_{-\infty}^{+\infty} 
\Bigg)^2
$$
having eliminated $-\int_{-\infty}^{+\infty}2x\mathrm{d}x$ because integral of a symmetrical function on both axes.

Since 
$\big[ x^4 \big]_{-\infty}^{+\infty}
\equiv \Bigg( 
\big[ x^2 \big]_{-\infty}^{+\infty} 
\Bigg)^2
$:

$$
\frac{\mu_4}{\sigma^4} - 3
= \frac{\big[ x^4 \big]_{-\infty}^{+\infty}}{\Bigg( 
\big[ x^2 \big]_{-\infty}^{+\infty} 
\Bigg)^2} - 3
= 1 - 3 = - 2
